```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
```

```python
df = pd.read_csv("resources/cleaned_data_t.csv")
```

```python
df
```

```python
group = df.groupby("Participant").mean()
```

```python
slope, intercept, r_value, p_value, std_err = st.linregress(group["Fatigue"],group["Active"])
r_2 = r_value**2
print(p_value.round(4))
y = group["Fatigue"]*slope + intercept
plt.scatter(group["Fatigue"],group["Active"])
plt.plot(group["Fatigue"], y , color = "red")
```

```python
# Correlation between Fatigue and PSMU
slope, intercept, r_value, p_value, std_err = st.linregress(group["Fatigue"],group["PSMU"])
r_2 = r_value**2

y = group["Fatigue"]*slope + intercept
plt.scatter(group["Fatigue"],group["PSMU"])
plt.plot(group["Fatigue"], y , color = "red")
plt.title("Fatigue vs PSMU")
plt.xlabel("PSMU")
plt.ylabel("Fatigue")

print(f"The p value for Fatigue and PSMU is {p_value.round(3)}")
print(f"The r squared value for Fatigue and PSMU is {r_2.round(3)}")
corr = round(st.pearsonr(group["Fatigue"],group["PSMU"])[0],2)
print(f"The correlation between Faitgue and PSMU is {corr}")

plt.savefig("Fatigue vs PSMU")
```
